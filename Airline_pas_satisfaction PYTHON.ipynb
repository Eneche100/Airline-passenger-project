```python
import pandas as pd
import numpy as np
```

```python
Airline_pas_satisfaction= pd.read_csv(r"C:\Users\LENOVO\Documents\airline_passenger_satisfaction.csv")
Airline_pas_satisfaction
```

```python
nan_values=Airline_pas_satisfaction.isna().sum()
nan_values
```

```python
Airline_pas_satisfaction["Arrival Delay"].unique()
```

```python
Airline_pas_satisfaction["Arrival Delay"].mean()
```

```python
Airline_pas_satisfaction["Arrival Delay"].isnull().sum()
```

```python
mean_values= Airline_pas_satisfaction["Arrival Delay"].mean()
mean_values
```

```python
Airline_pas_satisfaction["Arrival Delay"].fillna(mean_values, inplace=True)
Airline_pas_satisfaction
```

```python
Airline_pas_satisfaction["Arrival Delay"].isnull().sum()
```

```python
Airline_pas_satisfaction= Airline_pas_satisfaction.replace("Neutral or Dissatisfied", "Dissatisfied")
Airline_pas_satisfaction
```

```python
bins= [0, 20, 40, 60, 85 ]
labels= ["<20", "20-39", "40-59","60-85"]

Airline_pas_satisfaction["Age Group"] = pd.cut(Airline_pas_satisfaction["Age"], bins=bins, labels=labels, right=False)
Airline_pas_satisfaction
```

```python
Airline_pas_satisfaction["Satisfaction code"]= Airline_pas_satisfaction["Satisfaction"].copy()
```

```python
Airline_pas_satisfaction
```

```python
Airline_pas_satisfaction["Satisfaction code"]= Airline_pas_satisfaction["Satisfaction code"].replace({"Satisfied": 5, "Dissatisfied": 1})
```

```python
Airline_pas_satisfaction
```

```python
Air_pass_satis_corr= pd.DataFrame(Airline_pas_satisfaction, columns=["On-board Service","Seat Comfort","Cleanliness","Food and Drink","In-flight Service","In-flight Entertainment","Baggage Handling", "Satisfaction code"])
Air_pass_satis_corr
```

```python
correlation_matrix=Air_pass_satis_corr.corr()
correlation_matrix
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap="Blues", linewidths=0)
plt.title("correlation between different satisfaction factors")
plt.show()
```
